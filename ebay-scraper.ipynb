```python
import numpy
import bs4
from urllib.request import urlopen as ureq
from bs4 import BeautifulSoup as soup
my_url='https://www.ebay.com/b/DSLR-Cameras/31388/bn_732?LH_BIN=1&rt=nc'
uClient=ureq(my_url)
```

```python
page_html=uClient.read()
uClient.close()
page_soup=soup(page_html, "html.parser")
page_soup.h1
bizattri=page_soup.findAll("div", {"class":"s-item__info clearfix"})
```

```python
def isfloat(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
```

```python
priceList = []
nameList=[]
ratingList=[]
for container in bizattri:
    name=container.h3.text
    nameList.append(name)
    
    ratings=container.findAll("span",{"class": "clipped"})
#     print(len(ratings))
    if len(ratings) > 0:
        if isfloat(ratings[0].text.split()[0]):
            ratingList.append(float(ratings[0].text.split()[0]))
        else:
            ratingList.append(None)
    else:
        ratingList.append(None)
    
    
    find_price=container.findAll("span",{"class": "s-item__price"})
    priceList.append(float(find_price[0].text[1:].replace(',', '')))
    
    
```

```python
print(ratingList)
print(priceList)
print(nameList)
```

```python
Dict= {'Product':nameList, 'Price':priceList, 'Ratings':ratingList}
```
